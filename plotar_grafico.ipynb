```python
import matplotlib.pyplot as plt
import random as rand

ANOS = range(2004, 2018)

years = []
values = []

for year in ANOS:
    years.append(year)
    value = round(rand.random() * 100, 2)
    values.append(value)
    
print(years, values)
```

```python
plt.grid(True, linestyle="--")
plt.title("Valores por Ano\n", fontsize="20")
plt.xlabel("Ano", fontsize="16")
plt.ylabel("Em milhões de R$", fontsize="16")
plt.plot(ANOS, values, label='Valores Gastos')
legend = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, fontsize=12, borderaxespad=0.)
legend.get_frame().set_facecolor('#ffffff')
plt.show()
```
